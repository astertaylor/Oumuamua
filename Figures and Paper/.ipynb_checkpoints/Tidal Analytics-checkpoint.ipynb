```python
def analytics(r,theta,a,c,rho0):
    G=6.67e-8
    M=2e33
    alpha=2.236e26
    
    numer=r*(a**2-c**2)*np.sin(2*theta)
    denom=(r**2-(a**2-c**2)*np.cos(theta)**2)**2
    
    return(alpha*a*a*c*rho0*numer/denom)
```
